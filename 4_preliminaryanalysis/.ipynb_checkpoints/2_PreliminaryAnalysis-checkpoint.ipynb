<h3 style="background-color:#616161;color:white">1. Listening habits</h3>

<h4>How many tracks do people listen to in a day?</h4>

```python
sqlStr ='SELECT Cast(substr(userID,-5) as integer) as user,date(PlayedTimestamp),count(*) as NumOfPlays, count(Distinct trackID) as NumOfTracks from tblInputData group by userID, date(PlayedTimestamp)'

res = pd.read_sql_query(sqlStr, db)
```

```python
res[0:10]
```

```python
from bokeh.sampledata.autompg import autompg as df

p = Scatter(res, x='NumOfPlays', y='NumOfTracks', color='user', title="Daily listening habits of users",
            xlabel="NumOfPlays", ylabel="NumOfTracks")

output_notebook()
show(p)
```
